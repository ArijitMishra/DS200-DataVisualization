```python
import pandas as pd
```

```python
df = pd.read_excel('datafile.ods', engine='odf')
```

```python
df.head()
```

```python
df.keys()
```

```python
import matplotlib.pyplot as plt

#Scatter plot
f = plt.figure(figsize=(10,5))
plt.scatter(df['Under Trial Prisoner'],df['Convict Prisoner'])
plt.xlabel('Under Trial')
plt.ylabel('Convicted')
plt.show()
```

```python
#Box plot
data = [df['Under Trial Prisoner'],df['Convict Prisoner']]

fig = plt.figure(figsize=(9,7))
ax = fig.add_axes([0,0,1,1])
bp = ax.boxplot(data)
ax.set_xticklabels(['Under Trial','Convicted'])
plt.show()
```

```python
#bar plot
fig = plt.figure(figsize=(15,5))

plt.bar(df['Sl. No.'],df['Under Trial Prisoner'])
plt.xlabel("Prison Id ")
plt.ylabel('No of prisoners under trial')
plt.xticks(df['Sl. No.'])
plt.show()
```

```python
#bar plot
fig = plt.figure(figsize=(15,5))

plt.bar(df['Sl. No.'],df['Convict Prisoner'])
plt.xlabel("Prison Id ")
plt.ylabel('No of prisoners Convicted')
plt.xticks(df['Sl. No.'])
plt.show()
```
